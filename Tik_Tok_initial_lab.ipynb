```python
# What are the different values for claim status and how many of each are in the data?
tik_data['claim_status'].unique() #Distinct claim status values
tik_data['claim_status'].value_counts(dropna = False)
```

**Observation of values shown:** The counts of each claim status are nearly balanced. There are 2 distinct values recorded for claim status: claim, opinion. There are a small amount of null entry records recorded, which can indicate a claim being submitted without an actual response.

```python
# What is the average view count of videos with "claim" status?
claim_mean = tik_data[tik_data['claim_status'] == 'claim']['video_view_count'].mean()
claim_median = tik_data[tik_data['claim_status'] == 'claim']['video_view_count'].median()
print(f'The mean view count of videos with "claim" is: {claim_mean:.2f}')
print(f'The median view count of videos with "claim" is: {claim_median:.2f}')
```

```python
# What is the average view count of videos with "opinion" status?
opinion_mean = tik_data[tik_data['claim_status'] == 'opinion']['video_view_count'].mean()
opinion_median = tik_data[tik_data['claim_status'] == 'opinion']['video_view_count'].median()
print(f'The mean view count of videos with "opinion" is: {opinion_mean:.2f}')
print(f'The median view count of videos with "opinion" is: {opinion_median:.2f}')
```

**Observation of Mean and Median within each claim category:** The mean and the median within each claim category are close to one another, but there is a vast discrepancy between view counts for videos labeled as claims and videos labeled as opinions.

```python
# Get counts for each group combination of claim status and author ban status
tik_data.groupby(['claim_status', 'author_ban_status']).count()[['#']]
```

There are more claim videos with banned authors than there are opinion videos with banned authors. 
This could mean:
* Claim videos are more strictly policed than opinion videos
* Authors must comply with a stricter set of rules if they post a claim than if they post an opinion

It should be noted that there's no way of knowing if claim videos are inherently more likely than opinion videos to result in author bans, or if authors who post claim videos are more likely to post videos that violate terms of service.

Finally, while you can use this data to draw conclusions about banned/active authors, you cannot draw conclusions about banned videos. There's no way of determining whether a particular video _caused_ the ban, and banned authors could have posted videos that complied with the terms of service.

```python
# What's the median video share count of each author ban status?
tik_data.groupby(['author_ban_status']).median(numeric_only=True)[['video_share_count']]
```

Banned authors have a median share count that's 33 times the median share count of active authors! Explore this in more depth.

```python
### group the data by author_ban_status, then get the count, mean, 
# and median of each of the following columns: video_view_count, video_like_count, video_share_count
tik_data.groupby(['author_ban_status']).agg({
    'video_share_count': ['count','mean','median'],
    'video_like_count': ['count','mean', 'median'],
    'video_view_count': ['count','mean', 'median']
})
```

**The number of views, likes, and shares for banned authors compared to active authors:**

* Banned authors and those under review get far more views, likes, and shares than active authors.
* In most groups, the mean is much greater than the median, which indicates that there are some videos with very high engagement counts.

```python
#Create three new columns to help better understand engagement rates
```

```python
#likes per view column
tik_data['likes_per_view'] = tik_data['video_like_count']/tik_data['video_view_count']
#comments per view column
tik_data['comments_per_view'] = tik_data['video_comment_count']/tik_data['video_view_count']
#Shares per view column
tik_data['shares_per_view'] =tik_data['video_share_count']/tik_data['video_view_count']
```

```python
tik_data[['video_view_count','likes_per_view','comments_per_view', 'shares_per_view']].head()
```

```python
#Compile count, mean, and median for new columns based on claim status and author ban status
tik_data.groupby(['claim_status','author_ban_status']).agg({
    'likes_per_view': ['count', 'mean', 'median'],
    'comments_per_view': ['count', 'mean', 'median'],
    'shares_per_view': ['count', 'mean', 'median']    
})
```

**How does the data for claim videos and opinion videos compare or differ?**


We know that videos by banned authors and those under review tend to get far more views, likes, and shares than videos by non-banned authors. However, *when a video does get viewed*, its engagement rate is less related to author ban status and more related to its claim status.

Also, we know that claim videos have a higher view rate than opinion videos, but this tells us that claim videos also have a higher rate of likes on average, so they are more favorably received as well. Furthermore, they receive more engagement via comments and shares than opinion videos.

Note that for claim videos, banned authors have slightly higher likes/view and shares/view rates than active authors or those under review. However, for opinion videos, active authors and those under review both get higher engagement rates than banned authors in all categories.
